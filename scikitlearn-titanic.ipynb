```python
import torch
import numpy as np
import pandas as pd
 
```

```python
df_train=pd.read_csv('train.csv')
df_test=pd.read_csv('test.csv')
```

```python
df_train
```

```python
df_test
```

```python
df_train.isnull().sum()
```

```python
df_test.isnull().sum()
```

```python
mean_age=df_train['Age'].mean()
df_train['Age'].replace(np.nan,mean_age, inplace=True)
df_train['Cabin'].fillna('U',inplace=True)
df_train['Embarked'].dropna(inplace=True)
df_train.isnull().sum()
```

```python
mean_age=df_test['Age'].mean()
df_test['Age'].replace(np.nan,mean_age, inplace=True)
df_test['Cabin'].fillna('U',inplace=True)
df_test['Embarked'].dropna(inplace=True)
df_test.isnull().sum()
```

```python
df_train.drop(columns=['PassengerId','Name','Ticket','Fare','Cabin', 'Embarked'],inplace=True)
df_train
```

```python
df_test.drop(columns=['PassengerId','Name','Ticket','Fare','Cabin', 'Embarked'],inplace=True)
df_test
```

```python
df_train['Age'].max()
```

```python
df_train['Sex'].replace(['male','female'],[1,0], inplace=True)
df_test['Sex'].replace(['male','female'],[1,0], inplace=True)

df_train['Age']=df_train['Age']/df_train['Age'].max()
df_test['Age']=df_test['Age']/df_test['Age'].max()
df_test
```

```python
X_train=df_train.iloc[:,1:]
Y_train=df_train.iloc[:,:1]
X_test=df_test
X_train.shape
```

```python
X_train=np.array(X_train).astype('float32')
X_test=np.array(X_test).astype('float32')
Y_train=np.array(Y_train).astype('float32')
Y_train=np.ravel(Y_train)
Y_train
#X_test.shape
#X_test[0].reshape(1,-1)
```

```python
#https://towardsdatascience.com/logistic-regression-using-python-sklearn-numpy-mnist-handwriting-recognition-matplotlib-a6b31e2b166a
#https://stackoverflow.com/questions/34165731/a-column-vector-y-was-passed-when-a-1d-array-was-expected

from sklearn.linear_model import LogisticRegression
logisticRegr = LogisticRegression()
logisticRegr.fit(X_train, Y_train)
predicciones=logisticRegr.predict(X_test)
predicciones
```

```python
x_test=[3,0,0.28947368,1,1]
x_test=np.reshape(x_test,(1,5))
pred=float(logisticRegr.predict(x_test))
pred
```

```python
pkl_filename="kaggle_titanic_model.pkl"
with open(pkl_filename, 'wb') as file:  
    pickle.dump(logisticRegr, file)
```
